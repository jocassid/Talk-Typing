## What Do Type Annotations Do?

1. Avoid type mistakes (i.e. passing the wrong type of arguments to a function)
2. Make code more readable
3. Tools such as PyCharm or VSCode will show function/method signature if you
hover over the call.  If you add type annotations of the arguments and return
value(s), the types will be shown in the pop-up
4. When building a new function, it helps if you can define the type of data
going in (arguments) and out (return values).

## What Type Annotations DON'T Do?

Type annotations don't make Python statically typed.  Any variable may hold any type of value.

## Simple Types

```python
# numeric types
an_int: int = 10
a_float: float = 10.0
a_complex: complex = 5 + 10j

# boolean type
a_bool: bool = True

# text and binary sequence types
a_str: str = 'foo'
some_bytes: bytes = b'foo'
a_bytearray: bytearray = bytearray()
a_memory_view: memoryview = memoryview(b'foo')
```

## Sequences, Sets, and Mappings

```python
# sequence, set, and mapping types
# We're not saying anything about the type of values within sequence/mapping
a_list: list = []
a_tuple: tuple = tuple()

a_set: set = set()
a_frozenset: frozenset = frozenset()

a_dict: dict = {}
```

## Sequence, Sets, Mappings, and their contents

```python
list_of_ints: list[int] = [1, 2, 3]
a_tuple: tuple[bool, int] = (True, 5)

a_set: set[int] = {2, 3, 5, 7}
a_frozenset: frozenset[float] = frozenset([0.5, 0.25, 0.125])

a_dict: dict[str, int] = {'Ohio': 1803, 'New York': 1788}
```

## Sequences, Sets, and Mappings Prior to Python 3.9

Prior to Python 3.9, you need to use the type aliases from the `typing` module.

```python
list_of_ints: List[int] = [1, 2, 3]
a_tuple: Tuple[bool, int] = (True, 5)

a_set: Set[int] = {2, 3, 5, 7}
a_frozenset: FrozenSet[float] = frozenset([0.5, 0.25, 0.125])

a_dict: Dict[str, int] = {'Ohio': 1803, 'New York': 1788}
```

## Tuples of Varying Size

```python
IntTuple = tuple[int, ...]
EmptyTuple1 = tuple[()]

a: IntTuple = (1, 2)
b: IntTuple = (3, 4, 5)
c: IntTuple = ()
d: IntTuple = (1, 2, 'foo')  # TYPE ERROR

e: EmptyTuple1 = ()
f: EmptyTuple1 = (1, 2, 3)  # TYPE ERROR
```

## tuple[int, str, ...]

Is it a tuple consisting of ...
* int, str, one or more str
* int, str repleated

Actaully, it's the same as `tuple[int, str, Any]`


## Union

If a variable can hold more than one type, use the `|` operator to build a union of types.

```python
numeric = int | float | complex | Fraction | Decimal

# For Python < 3.10
# from typing import Union
# numeric = Union[int, float, complex, Fraction, Decimal]

def add_numeric(a: numeric, b: numeric) -> numeric:
    return a + b
```


## Optional

Use `| None` (or `Optional[TYPE]` in Python < 3.10) to indicate that a variable may be None

```python3
def greeting(name: str | None = None):
    name = name or 'stranger'
    print(f'Hello, {name}!')

def greeting_before_py3_10(name: Optional[str] = None):
    greeting(name)
```

## Any

Use `Any` if you don't know what type of data might be stored in a given variable

```python
def main():
    cache: dict[str, Any] = {
        'user_profile': {
            'username': 'carol',
            'theme_default': 'dark',
        },
        'cart': [
            {'product_id': 123, 'quantity': 2},
            {'product_id': 456, 'quantity': 1},
        ]
    }
```

## History of Type Annotations

### Python 3.5

* Introduction of standard library `typing`

### Python 3.6

From "[What's new in Python 3.6](https://docs.python.org/3/whatsnew/3.6.html#typing)"

> * "Improved support for generic type aliases.  For example `Dict[str, Tuple[S, T]]` is now a valid type annotation"
> * The `typing.ContextManager` class has been added for representing > `contextlib.AbstractContextManager`.
> * The `typing.Collection` class has been added for representing > `collections.abc.Collection.`
> * The `typing.ClassVar` type construct has been added to mark class > variables. As introduced in PEP 526, a variable annotation wrapped in ClassVar indicates that a given attribute is intended to be used as a class variable and should not be set on instances of that class.
> * A new `TYPE_CHECKING` constant that is assumed to be True by the static type checkers, but is False at runtime.
> * A new `NewType()` helper function has been added to create lightweight distinct types for annotations

### Python 3.6.1

* `typing.NamedTuple` supports default values, methods, and docstrings


### Python 3.9

* Eliminated the need for `Dict`, `List`, `Set`, `FrozenSet`, `Tuple`, `Type`
aliases.  Can now use `list[int]` instead of `List[int]`
* Eliminated the need for aliases to types in other standard library modules:
  * `collections`: `DefaultDict`, `OrderedDict`, `ChainMap`, `Counter`, `Deque`
  * `re`: `Match`, `Pattern`
* `typing.NamedTuple` not a function rather than class, but still can be used like `class MyTuple(NamedTuple)`

### Python 3.10

* `Union[X, Y]` may now be written as `X | Y`

### Python 3.11

* Added support for `NamedTuple` generics.

### Python 3.14

* Using `super()` and `__class__` in subclasses of `NamedTuple` not supported and will cause a `TypeError`
