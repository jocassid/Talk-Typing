## Introduction

### What Do Type Annotations Do?

1. Avoid type mistakes (i.e. passing the wrong type of arguments to a function)
2. Make code more readable
3. Tools such as PyCharm or VSCode will show function/method signature if you
hover over the call.  If you add type annotations of the arguments and return
value(s), the types will be shown in the pop-up
4. When building a new function, it helps if you can define the type of data
going in (arguments) and out (return values).

### What Type Annotations DON'T Do?

Type annotations don't make Python statically typed.  Any variable may hold any type of value.

### Union

If a variable can hold more than one type, use the `|` operator to build a union of types.

````python
from decimal import Decimal
from fractions import Fraction

numeric = int | float | complex | Fraction | Decimal

# For Python < 3.10
# from typing import Union
# numeric = Union[int, float, complex, Fraction, Decimal]

def add_numeric(a: numeric, b: numeric) -> numeric:
    return a + b
````

### Optional

Use `| None` (or `Optional[TYPE]` in Python < 3.10) to indicate that a variable may be None

````python
from typing import Optional

def greeting(name: str | None = None):
    name = name or 'stranger'
    print(f'Hello, {name}!')

def greeting_before_py3_10(name: Optional[str] = None):
    greeting(name)
````

### Building Nested Types

When building complex types, it helps to associate a component with a name.  These variables containing a type aren't to be confused with the TypeVar in the next section.

````python
FirstLastName = tuple[str, str]
CourseEnrollment = dict[str, list[FirstLastName]]

enrollment: CourseEnrollment = {
    'Math': [('Alice', 'Alison'), ('Bob', 'Bobinski')],
}
````

### TypeVar

`TypeVar`s are placeholders to allow you to define generic functions and classes

````python
from copy import deepcopy
from typing import TypeVar

T = TypeVar('T')

def duplicate(value_in: T) -> T:
    return deepcopy(value_in)

l1: list[int] = [2, 3]
l2: list[int] = duplicate(l1)
assert l1 == l2 and l1 is not l2

t1: tuple[int, int] = [3, 4]
t2: tuple[int, int] = duplicate(t1)
assert t1 == t2 and t1 is not t2
````

### Any

### Classes

Classes may be used in type annotations

````python
from dataclasses import dataclass

@dataclass
class Point:
    x: int = 0
    y: int = 0
    
    def distance(self, other: 'Point') -> float:
        return sqrt((other.x - self.x)**2 + (other.y - self.y)**2)

p1: Point = Point(1, 2)
p2: Point = (1, 2)  # TYPE ERROR
````

### Classes - Of the non-dataclass variety

````python
from math import sqrt

class Line:
    def __init__(self, start: Point, end: Point):
        self.start: Point = start
        self.end: Point = end

    @property
    def length(self) -> float:
        return self.start.distance(self.end)
````

### Callable

````python
from typing import Callable

# Callable[[arg_type1, arg_type2], return_type]
Filter = Callable[[str], bool]

def lowercase_filter(text_in: str) -> bool:
    return text_in == text_in.lower()

def less_than_5_letters(text_in: str) -> bool:
    return len(text_in) < 5

filters: list[Filter] = [lowercase_filter, less_than_5_letters]
````

### Callable acccept any number and types of arguments

````python
from typing import Callable

IntFunction = Callable[..., int]

def add_two(a: int, b: int) -> int: return a + b
def add_three(a: int, b: int, c: int) -> int: return a + b + c
def str_len(a: str) -> int: return len(a.strip())

int_functions: list[IntFunction] = [
    add_two,
    add_three,
    str_len,
]
````

### Callable - Complex Args - Protocol

`Callable` doesn't let specify types for `*args` and `**kwargs`.  Create a `Protocol` with a `__call__` method

````python
from typing import Protocol

class Event: ...

class EventHandler(Protocol):
    def __call__(self, event: Event, *args: Any, log_event: bool) -> bool: ...

def handle_click(event: Event, *args, log_event: bool) -> bool:
    ...

handler: EventHandler = handle_click
````

### Create Your Own Generic Types

````python
from random import randint
from typing import TypeVar

T = TypeVar('T')

class FirstInRandomOut[T]:
    def __init__(self, contents: list[T] | None = None):
        self.contents: List[T] = contents
    def append(self, item: T) -> None:
        self.contents.append(item)
    def pop(self) -> T:
        if len(self.contents) < 1:
            raise IndexError('pop from empty')
        return self.contents.pop(
            randint(0, len(self.contents) - 1)
        )

firo_queue: FirstInRandomOut[int] = FirstInRandomOut([1, 2, 3, 5, 8, 13])
print(firo_queue.pop())
print(firo_queue.pop())
````

### Generators

## History of Type Annotations

### Python 3.5

* Introduction of standard library `typing`

### Python 3.6

From "[What's new in Python 3.6](https://docs.python.org/3/whatsnew/3.6.html#typing)"

> * "Improved support for generic type aliases.  For example `Dict[str, Tuple[S, T]]` is now a valid type annotation"
> * The `typing.ContextManager` class has been added for representing > `contextlib.AbstractContextManager`.
> * The `typing.Collection` class has been added for representing > `collections.abc.Collection.`
> * The `typing.ClassVar` type construct has been added to mark class > variables. As introduced in PEP 526, a variable annotation wrapped in ClassVar indicates that a given attribute is intended to be used as a class variable and should not be set on instances of that class.
> * A new `TYPE_CHECKING` constant that is assumed to be True by the static type checkers, but is False at runtime.
> * A new `NewType()` helper function has been added to create lightweight distinct types for annotations

### Python 3.6.1

* `typing.NamedTuple` supports default values, methods, and docstrings


### Python 3.9

* Eliminated the need for `Dict`, `List`, `Set`, `FrozenSet`, `Tuple`, `Type`
aliases.  Can now use `list[int]` instead of `List[int]`
* Eliminated the need for aliases to types in other standard library modules:
  * `collections`: `DefaultDict`, `OrderedDict`, `ChainMap`, `Counter`, `Deque`
  * `re`: `Match`, `Pattern`
* `typing.NamedTuple` not a function rather than class, but still can be used like `class MyTuple(NamedTuple)`

### Python 3.10

* `Union[X, Y]` may now be written as `X | Y`
* `Callable` now supports `ParamSpec` and `Concatenate`

### Python 3.11

* Added support for `NamedTuple` generics.

### Python 3.14

* Using `super()` and `__class__` in subclasses of `NamedTuple` not supported and will cause a `TypeError`
